```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
N = []
for i in range(500):
    N.append(int(1*(i)+1))
    N.append(int(1*(i)+1))
file = open("test_vhod.txt","w")
for i in N:
    file.write(str(i)+"\n");
file.close()
```

```python
N_1 = [[],[]]
file = open("2DiscretS.txt","r")
for line in file:
    line = line.split()
    N_1[0].append(float(line[0]))
    N_1[1].append(float(line[1]))
file.close()
N_1 = np.array(N_1)
```

```python
plt.figure(figsize=(10, 7))
plt.plot(N_1[0],N_1[1],"o")
plt.suptitle("Зависимость вероятности попадания в число от отношения 1 к величине диапазона")
plt.xlabel("Отношение 1 к величине диапазона")
plt.ylabel("Вероятность попадания")
```
